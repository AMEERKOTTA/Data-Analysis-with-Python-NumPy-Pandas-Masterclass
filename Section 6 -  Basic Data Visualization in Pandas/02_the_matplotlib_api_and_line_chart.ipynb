### THE MATPLOTLIB API

+ Pandas uses the matplotlib API to plot and visualize the data.
+ This is an Integration with the main matplotlib library.

THE PLOT METHOD.

+ df.plot() will plot you a line chart.
+ row index as the x-axis.
+ and plotting each numerical column as a seperate series on the y-axis.
+ you can change the x-axis by setting a different index or using the x argument.
+ you can select the series by using .loc[] argument.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("stores.csv")
df.head()
```

```python
## plot the df, this will only plot the numerical columns.
## here the numerical columns are store_nbr and cluster
df.plot()
```

```python
transactions = pd.read_csv("transactions.csv")
transactions
## plot the line chart based on the date on x-axis.
transactions.set_index("date").plot()
```

```python
## plot the line chart based on the date on x-axis.
transactions.plot(x = "date")
```

```python
transactions.head(20)
```

```python
## plotting the first 20 values in the dataset.
transactions.set_index("date").iloc[:20, :].plot()
```

```python
## plotting the first 20 values in the dataset., plotted only transactions
transactions.set_index("date").loc["2013-01-01":"2013-01-02", "transactions"].plot()
```

```python
## plotting the first 20 values in the dataset., plotted only store_nbr
transactions.set_index("date").loc["2013-01-01":"2013-01-02", "store_nbr"].plot()
```

```python
## also we can plot like 
transactions.plot(x = "date", y = "transactions")
```

```python
transactions = pd.read_csv("transactions.csv")
transactions.head()
```

```python
## plot 
transactions.plot()
```

```python
transactions.set_index("date").loc[:"2013-01-05","transactions"].plot()
```

```python
transactions.query("store_nbr == 2").set_index("date")
```

```python
transactions.query("store_nbr == 44").set_index("date").loc[:"2013-01-10","transactions"].plot()
```

```python
transactions.query("store_nbr == 2").set_index("date").loc[:"2013-01-10","transactions"].plot()
```

```python
(transactions
    .query("store_nbr in ['44',23]")
    .pivot_table(index = "date", columns = "store_nbr")
    .droplevel(0, axis = 1)
    .plot()
);
```
