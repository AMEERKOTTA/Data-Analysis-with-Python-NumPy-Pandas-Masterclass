### GROUPBY METHOD

+ Grouping Dataframes will allow you to aggregate the data at a different level.
+ To group data, use the `.groupby()` method, and specify a column to groupby.
+ The grouped column becomes the index by default.

```python
## reading teh retail dataset
retail = pd.read_csv("retail_2016_2017.csv")
retail.head()
```

```python
## groupby "family"
small_retail = retail.groupby("family")
small_retail

## perform some kind of aggregation on some column
retail.groupby("family")["sales"].sum().head()
```

```python
## perform some kind of aggregation on some column
retail.groupby("family")[["sales"]].mean().head()
```

```python
df.head()
df.groupby(["HomeTeam"])[["Home Goals"]].sum()
```

```python
df.groupby(["HomeTeam"])[["Home Goals"]].sum().sort_values(by = "Home Goals", ascending = False).head()
```

```python
df.groupby(["AwayTeam"])[["Away Goals"]].sum().sort_values(by = "Away Goals", ascending = False).head()
```

```python
## reading the transactions data
transactions = pd.read_csv("transactions.csv")
transactions.head()
transactions.groupby("store_nbr")[["transactions"]].sum().sort_values(by = "transactions", ascending = False).head(10)
```

### Grouping by Multiple Columns

+ This can be done by Passing the list of Columns to the groupby method.
+ This create a multi index object.

```python
retail = pd.read_csv("retail_2016_2017.csv")
retail.head()
```

```python
retail.groupby(["family","store_nbr"], as_index = False)[["sales"]].sum().loc[:100].query("family == 'AUTOMOTIVE'")
```

```python
## transactions
transactions
## adding a column "month" to transactions
tr_df = transactions.assign(
    month = transactions["date"].str[5:7].astype("int")
)
tr_df.head()
```

```python
tr_df.groupby(["store_nbr","month"])[["transactions"]].sum().sort_values(by = ["month","transactions"], ascending = [True, False])
```
