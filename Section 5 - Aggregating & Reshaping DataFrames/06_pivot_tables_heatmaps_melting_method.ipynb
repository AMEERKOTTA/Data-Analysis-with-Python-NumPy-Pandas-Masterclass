### PIVOT TABLES

+ they are great ways to create summary tables of data.
+ .pivot_tables() lets you create Excel style pivot table

pivot tables methods arguments...

1. `index` :-- returns a row index with distinct values from the specified column.
2. `columns` :-- returns a column index with a distinct values from the specified column.
3. `values` :-- the column or the columns to perform the aggregations on
4. `aggfunc` :-- defines the aggregation function, or functions to perform on the values.
5. `margin` :-- return row and column totals when True.


Multiple Aggregation Function in a Pivot Table.

+ Here we have to pass multiple aggregations like sum, mean, min, max or etc


**HeatMaps**

You can style a dataframe based on its value to create a heatmap.

+ `df.style.background_gradient()` to the df and add a cmap argument.

```python
import pandas as pd
import numpy as np
```

```python
retail = pd.read_csv("retail_2016_2017.csv")
retail.loc[:5]

retail_pivot = retail.pivot_table(
    index = "family",
    columns = "store_nbr",
    values = "sales",
    aggfunc = ["max"]
)
retail_pivot
```

```python
retail_pivot = retail.pivot_table(
    index = "family",
    columns = "store_nbr",
    values = "sales",
    aggfunc = ["max"],
    margins = True
)
retail_pivot.head()
```

```python
## multplile aggregations

retail_pivot = retail.pivot_table(
    index = "family",
    columns = "store_nbr",
    values = "sales",
    aggfunc = ["max", "min"],
    margins = True
)
retail_pivot.head()
```

### HEATMAPS

```python
retail_pivot = retail.pivot_table(
    index = "family",
    columns = "store_nbr",
    values = "sales",
    aggfunc = ["max"],
    margins = True
)
retail_pivot.style.background_gradient(cmap = "RdYlGn", axis = None)
```
