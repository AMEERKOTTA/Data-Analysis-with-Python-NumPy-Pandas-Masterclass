### TRANSFORMING DATAFRAMES

+ the .transform() method can be used to perform aggregations without reshaping.
+ this is useful to calculate group level statistics to perform row lwvwl analysis.

```python
import numpy as np
import pandas as pd
```

```python
premier_league = pd.read_csv("2021-2022.csv")
premier_league
pl_df = premier_league[["Date","Time","HomeTeam","AwayTeam","FTHG","FTAG"]]
pl_df.rename(columns = {"FTHG" : "HomeGoals", "FTAG" : "Away Goals"}, inplace = True)
pl_df
```

```python
pl_df.head()
```

```python
pl_df.assign(
    avg_team_goals = pl_df.groupby(["HomeTeam"])["HomeGoals"].transform("sum"),
    mean_team_goals = pl_df.groupby(["HomeTeam"])["HomeGoals"].transform("mean"),
    difference = lambda x : x["HomeGoals"] - x["mean_team_goals"]
)
```

```python
pl_df.groupby(["HomeTeam","AwayTeam"]).agg({'difference' : 'sum'}).sort_values(by = "difference", ascending = False)
```
