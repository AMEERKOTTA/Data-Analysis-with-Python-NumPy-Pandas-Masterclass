### FILTERING DATAFRAMES

+ You can filter the Dataframe by Passing the Filters(logical test) using .loc Accessor.

```python
import pandas as pd
import numpy as np
```

```python
retail = pd.read_csv("retail_2016_2017.csv")
retail

## filtering by date.
retail[retail["date"] == "2016-01-01"]
```

```python
## fiterng by store number
retail[retail["store_nbr"] == 1]
```

```python
## filtering by date and store number
retail[(retail["date"] == "2016-01-01") & (retail["store_nbr"] == 1)]
```

```python
retail[(retail["family"].isin(["AUTOMOTIVE","HOME CARE"])) & (retail["sales"] > 0) & (retail["date"] == "2016-01-01")]
```

#### THE .QUERY METHOD

This will allow you to put any number of filters by combining the and, or etc.

```python
retail.query('family in ["CLEANING", "DIARY"] and sales > 0 and date in ["2016-01-01"]')
```

```python
retail.query('family in ["CLEANING"] and sales >= 0')
```

```python
transactions = pd.read_csv("transactions.csv")
transactions
```

```python
## filter by transactions > 2000
transactions[transactions["transactions"] > 2000]
```

```python
transactions[(transactions["transactions"] > 2000) & (transactions["store_nbr"] == 25)]
```

```python
transactions.query("store_nbr in [31, 25] and transactions < 2000").sum().iloc[2]
```
