### SORTNG DATAFRAMES

+ There is a Method to Sort Dataframes by its Index using `.sort_index().`
+ This sorts row (axis = 0 by default) and can specify (axis = 1 to get sorted by columns)

+ You can sort the Dataframes by Values by `.sort_values()` method.

```python
import numpy as np
import pandas as pd
```

```python
retail = pd.read_csv("retail_2016_2017.csv")
retail.head()

condition = retail.family.isin(["BEVERAGES","DELI","DIARY"])
df = retail[condition]
df
```

```python
## sorting
df.sort_index()
```

```python
## sorting
df.sort_index(ascending=False)
```

```python
## sorting by columns 
df.sort_index(axis = 1)
```

```python
### sort by values
df.sort_values("id")
```

```python
df.sort_values("family")
```

```python
df.sort_values("sales")
```

```python
df[["store_nbr", "sales"]].sort_values(["sales","store_nbr"], ascending=[True, False])
```

```python
transactions = pd.read_csv("transactions.csv")
transactions.head()
```

```python
## top five transactions.
df_top_5 = transactions.sort_values("transactions", ascending = False)
df_top_5.iloc[:5]
```
