### DATAFRAME BASICS

+ Dataframes are Pandas tables made up from Columns and Rows.
+ Each column is a Pandas Series that stores the same row Index.
+ The column headers works as column index that contains series names.


DATAFRAME PROPERTIES

+ shape :- no.of rows and columns 
+ index :- the row index (axis = 0) (integers)
+ columns :- the column index (axis = 1)
+ axes :- the row and column indices.
+ dtypes :- the data types in each series in a DataFrame.

### CREATING A DATAFRAME

+ You can create a Dataframe from Python dictionary or Numpy array.
+ You will more likely create a Dataframe by reading in a flat file like (csv, txt, tsv etc)
