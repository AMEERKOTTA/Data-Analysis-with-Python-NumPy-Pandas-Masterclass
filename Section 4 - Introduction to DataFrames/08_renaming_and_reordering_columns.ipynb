### RENAMING AND REORDERING COLUMNS

+ Rename the columns by assignments, `df.columns = ["column1","column2","column3"]`
+ Another methof is by using rename method,` df.rename(columns = {"column1":"column_one","column2":"column_two"})`

+ Reorder the columns by `.reindex` method.
+ `data.reindex(labels = ["column1","column2',"column3"], axis = 1)`

```python
import pandas as pd
import numpy as np
```

```python
oil_data = pd.read_csv("oil.csv")
oil_data

## rename the columns names
oil_data.columns = ["price", "oil"]
oil_data
```
